```python
# Code to clean the headers
import pandas as pd

# Assuming your DataFrame is named 'df'
# This code will rename the column headers

def clean_player_stats_headers(df):
    # Create a dictionary mapping old column names to new ones
    new_columns = {
        'Unnamed: 0_level_0_Player': 'Player',
        'Unnamed: 1_level_0_Nation': 'Nation',
        'Unnamed: 2_level_0_Pos': 'Position',
        'Unnamed: 3_level_0_Age': 'Age',
        'Unnamed: 4_level_0_MP': 'Matches_Played',
        'Playing Time_Starts': 'Starts',
        'Playing Time_Min': 'Minutes',
        'Playing Time_90s': 'Minutes_90s',
        'Performance_Gls': 'Goals',
        'Performance_Ast': 'Assists',
        'Performance_PK': 'Penalties_Scored',
        'Performance_PKatt': 'Penalties_Attempted',
        'Performance_CrdY': 'Yellow_Cards',
        'Performance_CrdR': 'Red_Cards',
        'Per 90 Minutes_Gls': 'Goals_Per90',
        'Per 90 Minutes_Ast': 'Assists_Per90',
        'Per 90 Minutes_G+A': 'GoalsAndAssists_Per90',
        'Per 90 Minutes_G-PK': 'NonPenaltyGoals_Per90',
        'Per 90 Minutes_G+A-PK': 'NonPenaltyGoalsAndAssists_Per90',
        'Unnamed: 21_level_0_Matches': 'Match_Details'
    }
    
    # Rename the columns
    df = df.rename(columns=new_columns)
    
    return df
```

```python
# Example usage
df_players_stats_clean = clean_player_stats_headers(df_players_stats)
df_players_stats_clean
```
