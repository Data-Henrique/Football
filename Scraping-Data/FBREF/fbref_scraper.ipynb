## FBREF - Real Madrid Champions League 2023/24

```python
import pandas as pd
```

## Scraping Champions League 2023 / 2024 Stats

```python
# Extracting a table with the results regarding group stages and different knockout phases
# The URL used is simply the page URL
# The second parameter comes from using the Inspect function on the webpage and inspecting the table
# The "id" is to target the correct table
# The data is retrieved as a list and we use [0] to get the table
df_cl = pd.read_html("https://fbref.com/en/comps/8/2023-2024/2023-2024-Champions-League-Stats",
                  attrs = {"id" : "results2023-202480_overall"})[0]
```

```python
# The df came with a lot of NaN values created by the table separators so let's drop it
df_cl = df_cl.dropna(subset = ['Rk'])
```

## Scraping Real Madrid Players 2023 / 2024 Stats
